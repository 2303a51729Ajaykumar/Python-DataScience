```python
import pandas as pd


products = pd.DataFrame({
    "product_id": [101, 102, 103, 104, 105],
    "product_name": ["Laptop", "Mobile", "Headphones", "Keyboard", "Monitor"],
    "category": ["Electronics", "Electronics", "Accessories", "Accessories", "Electronics"],
    "price": [75000, 25000, 2000, 1500, 12000]
})


sales = pd.DataFrame({
    "sale_id": [1, 2, 3, 4, 5, 6],
    "product_id": [101, 103, 102, 105, 101, 104],
    "quantity": [2, 5, 3, 1, 1, 4],
    "date": pd.to_datetime(["2025-09-01", "2025-09-01", "2025-09-02", "2025-09-02", "2025-09-03", "2025-09-03"])
})
```

```python
merged=pd.merge(products,sales,on="product_id")
```

```python
merged
```

```python
inner=pd.merge(products,sales,on="product_id",how="inner")
```

```python
outer=pd.merge(products,sales,on="product_id",how="outer")
```

```python
inner
```

```python
outer
```

```python

merged["revenue"] = merged["quantity"] * merged["price"]


total_revenue = merged["revenue"].sum()


average_price = merged["price"].mean()
```

```python
total_revenue
```

```python
average_price
```

```python
product_sales = merged.groupby("product_name").agg(
    total_quantity=("quantity", "sum"),
    total_revenue=("revenue", "sum"),
    avg_price=("price", "mean")
).reset_index()
```

```python
product_sales
```

```python
import pandas as pd

teachers = pd.DataFrame({
    "teacher_id": [1, 2, 3, 4],
    "teacher_name": ["Alice", "Bob", "Charlie", "David"],
    "department": ["Math", "CS", "Physics", "Math"]
})

courses = pd.DataFrame({
    "course_id": [101, 102, 103, 104, 105, 106],
    "course_name": ["Algebra", "Data Structures", "Quantum Physics",
                    "Calculus", "AI", "Mechanics"],
    "teacher_id": [1, 2, 3, 1, 2, 3],
    "hours_per_week": [4, 5, 6, 3, 4, 5]
})

```

```python
teachers
```

```python
courses
```

```python
inner_join = pd.merge(courses, teachers, on="teacher_id", how="inner")
print("---- INNER JOIN ----")
print(inner_join, "\n")
```

```python
left_join = pd.merge(courses, teachers, on="teacher_id", how="left")
print("---- LEFT JOIN ----")
print(left_join, "\n")
```

```python
avg_courses_per_teacher = (
    courses.groupby("teacher_id")["course_id"].count().mean()
)
```

```python
avg_courses_per_teacher
```

```python

teacher_workload = inner_join.groupby(
    ["teacher_id", "teacher_name"]
)["hours_per_week"].sum().reset_index(name="total_hours")

dept_workload = inner_join.groupby("department")["hours_per_week"].sum().reset_index()

print("---- Average Courses per Teacher ----")
print(avg_courses_per_teacher, "\n")

print("---- Teacher Workload ----")
print(teacher_workload, "\n")

print("---- Department Workload ----")
print(dept_workload, "\n")
```
