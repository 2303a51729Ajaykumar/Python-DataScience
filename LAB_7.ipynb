```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#loading data set to pandas dataframe
data=pd.read_csv('/content/weatherHistory.csv')
```

```python
data.head()
```

```python
#plotting using line chart
chart=plt.plot('Temperature (C)','Humidity',data=data)
plt.show()
```

```python
#plotting using line chart
data['Day'] = pd.to_datetime(data['Formatted Date'], utc=True).dt.day
df = data[['Day', 'Temperature (C)']].head(7)
display(df)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Day'], df['Temperature (C)'], marker='o', linestyle='-', linewidth=2, markersize=6)
plt.title("Temperature Variation")
plt.xlabel("Day")
plt.ylabel("Temperature (C)")
plt.grid(True)
plt.show()
```
